# Primers Selection (ver. 2022.11.26)

```python
import os, glob, shutil
import random
from tqdm.notebook import tqdm
from pathlib import Path
import muspy as mu
from mido import MidiFile
import pretty_midi as pr
import pickle
```

```python
load_dir="/data/data1/users/astais/Unprocessed-Datasets/"
datasets={'maestro-v3.0.0':load_dir+'maestro-v3.0.0/',
          'GiantMIDI-Piano':load_dir+'GiantMIDI-Piano/',
          'ailabs1k7':load_dir+'ailabs1k7/midi_analyzed/',
          'Rock-Piano-MIDI-Dataset':load_dir+'Rock-Piano-MIDI-Dataset/',
          'adl-piano-midi':load_dir+'adl-piano-midi/',
          'Los-Angeles-MIDI-Dataset-segment':load_dir+'Los-Angeles-MIDI-Dataset-segment/'
         }
save_dir="/data/data1/users/astais/Unprocessed-Datasets/"
num=100 # number of primers frome each dataset
```

```python
primers={}
for name,path in tqdm(datasets.items()): 
    # count number of midis in directory
    # mid_num=len(tqdm([f for f in Path(directory).rglob('*.mid*')]))
    file_list = list(Path(path).glob(f"**/*.mid*"))
    if not len(file_list):
        print("No files found.")

    # iterate over files in directory and randomly copy 
    for i in tqdm(range(num),desc="Choosing "+str(num)+" primer files"):
        while True: 
            try:
                # pick a random file from the file list
                rand = random.randint(0, len(file_list) - 1)
                file=str(file_list[rand])
                primers.update({name+'_'+str(i+1):file})

                # test that the file can be processed
                pr.PrettyMIDI(file)
                mu.from_mido(MidiFile(file), duplicate_note_mode='fifo')
                # copy file to primers folder
#                 shutil.copy(file,save_dir)
            except:            
                # Error happened, continue
                print("Error happened. File: "+file)
                del primers[name+'_'+str(i+1)]
                continue
            else:
                break

    print('Done!')
```

```python
# create a binary pickle file 
f = open(save_dir+"primers.pickle","wb")
# write the python object (dict) to pickle file
pickle.dump(primers,f)
# close file
f.close()
```
